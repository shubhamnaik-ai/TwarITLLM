```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
df = pd.read_csv(working_dir_path + "testing.csv")
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
def row_to_text(row):
    return " | ".join(f"{col}: {row[col]}" for col in df.columns)

row_texts = df.apply(row_to_text, axis=1).tolist()
```

```python
row_embeddings = model.encode(row_texts, convert_to_tensor=True, show_progress_bar=True)
```

```python
def query_log(prompt, top_k=5):
    prompt_embedding = model.encode(prompt, convert_to_tensor=True)
    hits = util.semantic_search(prompt_embedding, row_embeddings, top_k=top_k)[0]

    print(f"\nTop {top_k} matching rows for query: \"{prompt}\"\n")
    for i, hit in enumerate(hits):
        idx = hit['corpus_id']
        score = hit['score']
        print(f"[{i+1}] Score: {score:.4f}")
        print(row_texts[idx])
        print("—" * 80)
```

```python
query_log("what is the average time for charging vehicles between 45 to 55 SOC", top_k=100)
```
